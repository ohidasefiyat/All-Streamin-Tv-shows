```python
#CREATE A NEW EXCEL FILE WITH SHEETS REPRESENTING EACH OF THE YEARS OF MOVIES
#MOVIES RELEASEDE AND THE DATA AS ALL MOVIES RELEASED IN SAID YEAR

yearly_movies_grouped = all_tv_shows.groupby(by = "Year Released")
```

```python
group_keys = yearly_movies_grouped.groups.keys() #retrieve keys from groups
```

```python
with pd.ExcelWriter("netflix_by_years.xlsx") as writer:
    
    for key in group_keys:
    #subFrame.to_excel(writer, sheet_name = "subFrame1", index = False)  
    #subFrame2.to_excel(writer, sheet_name = "subFrame2", index = False)
    #print(key, "--> ", yearly_movies_grouped.get_group(key).shape)
        dataframe = yearly_movies_grouped.get_group(key)
        dataframe.to_excel(writer, sheet_name = f"{key}", index = False )
```
